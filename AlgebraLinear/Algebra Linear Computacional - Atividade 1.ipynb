## Questão 1

> Calcular o número de condição da matriz de Hilbert de tamanho 10, 100 e 1000.

A fórmula para o elemento $(i,j)$ é $$\frac{1}{i+j-1}$$

Criando a matriz de Hilbert:

```julia
function MatrixHilbert(n)
  A = zeros(n,n)
  for i = 1:n
    for j = 1:n
      A[i,j] = 1/(i+j-1)
    end
  end
      A
end;
```

```julia
H10 = MatrixHilbert(10)
H100 = MatrixHilbert(100)
H1000 = MatrixHilbert(1000);
```

Vamos usar a função `cond` do `julia`

```julia
cond(H10)
```

```julia
cond(H100)
```

```julia
cond(H1000)
```

## Questão 2

Resolver o sistema $$Ax = A_{[:,end]}$$

```julia
# H10 já está definida
H20 = MatrixHilbert(20)
H40 = MatrixHilbert(40);
```

Vamos criar uma função auxiliar que produz os vetores solução


```julia
xsolucao(n) = begin
  v = zeros(n)
  v[n] = 1
  v
end
```

Vamos criar os vetores solução para comparar com as soluções encontradas


```julia
xs10 = xsolucao(10)
xs20 = xsolucao(20)
xs40 = xsolucao(40);
```

Vamos resolver o sistema usando o famoso operador `backslash`


```julia
x10 = H10\H10[:,end]
x20 = H20\H20[:,end]
x40 = H40\H40[:,end];
```

Comparando as soluções encontradas

```julia
errorelativo(x0, x) = norm(x0-x)/norm(x)
```

```julia
errorelativo(x10, xs10)
```

```julia
errorelativo(x20, xs20)
```

```julia
errorelativo(x40, xs40)
```

Note que o erro relativo até que está controlado no caso $n=10$, mas explode
em $n=20$ e $n=40$

## Questão 4

```julia
function substituicaoReversa(U,b)
  m,n = size(U)
  nb = length(b)
  if m != n
    throw( DimensionMismatch("Matriz U não é quadrada ($m x $n)"))
  end
  if nb != n
    throw(DimensionMismatch("Lado direito não tem o tamanho correto: U($m x $n), b($nb)"))
  end
  x = zeros(n)
  x[n] = b[n]/U[n,n]
  for i = n-1:-1:1
    x[i] = (b[i] - U[i,i:end]'*x[i:end])/U[i,i]
  end
  x
end
```

```julia
A = [1 2; 0 4]
b = [1; 2]
```

```julia
substituicaoReversa(A,b)
```

```julia
n = 20
B = rand(n,n)
A = UpperTriangular(B);
b = A*ones(n);
```

```julia
x= substituicaoReversa(A,b)
```
